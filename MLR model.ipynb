```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Data Cleaning
#Data wrangling
#Train Test Split

X= df.drop(columns='sales')
y= df.sales
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
#Modelling

from sklearn.linear_model import LinearRegression

model= LinearRegression()

model.fit(X_train,y_train)

```

```python
model.intercept_
```

```python
model.coef_
```

```python
#predictions

train_prediction= model.predict(X_train)
```

```python
test_prediction= model.predict(X_test)
```

```python
#RMSE
from sklearn.metrics import mean_squared_error

test_RMSE = np.sqrt(mean_squared_error(y_test,test_prediction))
train_RMSE = np.sqrt(mean_squared_error(y_train,train_prediction))
```

```python
print (test_RMSE, train_RMSE)
```

```python
model.score(X_train,y_train)   #train R2
```

```python
model.score(X_test,y_test)   #test R2
```

```python
from sklearn.model_selection import cross_val_score

scores= cross_val_score(model,X,y,cv=5)
print(scores)
```

```python
scores.mean()
```

```python
#check assumptions

#1. Linearity of errors

test_res = y_test-test_prediction
```

```python
plt.scatter(y_test,test_res)
plt.xlabel('observed values')
plt.ylabel('fitted values')
plt.show()
```

```python
# linearity of errors test failed

#2.Normality of error

sns.displot(test_res, kde=True, bins=15)
plt.show()
```

```python
#it is not normal. It is left skewwed

#3 Equal Variance of error (Homoscadesicity)

plt.scatter(test_prediction,test_res, c='r')
plt.axhline(color='red', y=0)
plt.xlabel('fitted values')
plt.ylabel('residual')
plt.show()
```

```python
#it is HetroScadisicity as it does not have equal variance

#Variable Significance

import statsmodels.formula.api as smf

model1= smf.ols("y~X", data=df).fit()
```

```python
import statsmodels.formula.api as smf
modelinfuence= smf.ols(formula="sales~TV+radio+newspaper", data=df_new).fit()
modelinfuence.summary()
```

```python
#p value is still greatter than zero for newspaper

#Variance inflation factor (VIF)
VIFmodel=smf.ols('TV~radio+newspaper', data=df).fit()
```

```python
rsq_tv=smf.ols('TV~radio+newspaper', data=df).fit().rsquared

vif_tv = 1/(1-rsq_tv)

rsq_radio=smf.ols('radio~TV+newspaper', data=df).fit().rsquared

vif_radio = 1/(1-rsq_radio)

rsq_newspaper=smf.ols('newspaper~TV+radio', data=df).fit().rsquared

vif_newspaper = 1/(1-rsq_newspaper)
```

```python
#storing vif values in the dataframe

d1 = {'Variables':['TV','radio','newspaper'], 'VIF':[vif_tv,vif_radio,vif_newspaper]}
VIFframe=pd.DataFrame(d1)
VIFframe
```
